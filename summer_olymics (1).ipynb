```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('summer.csv')

df.head()
```

```python
df
```

 1. In how many cities Summer Olympics is held so far?


```python
c=0
print('names of cities the summer olympics held so far\n')
for City in df['City'].unique():
  print(City)
  c=c+1
  print('\n total no of cities summer olympics held',c)
```

2. Which sport is having most number of Gold Medals so far? (Top 5)   



```python
gold=df[df['Medal']=='Gold']

data=[]
for Sport in gold['Sport'].unique():
  data.append([sport,len(gold[gold['Sport']==sport])])

data=pd.DataFrame(data,columns=['Sport','Medals'])

data=data.sort_values(by='Medals',ascending=False).head()

data.plot(x='Sport',y='Medals',kind='bar',figsize=(8,6))
```

```python
data
```

4. Which player has won most number of medals? (Top 5)

```python
data=[]

for athlete in df['Athlete'].unique():
  data.append([athlete,len(df[df['Athlete']==athlete])])

data=pd.DataFrame(data,columns=['Athlete','medal'])

data=data.sort_values(by='medal',ascending=False).head()

data.plot(x='Athlete',y='medal',kind='bar',figsize=(10,6),color='red')
```

7. Which event is most popular in terms on number of players? (Top 5)

```python
data=[]

for event in df['Event'].unique():
  data.append([event,len(df[df['Event']==event])])

data=pd.DataFrame(data,columns=['Event','Athlete'])

data=data.sort_values(by='Athlete',ascending='False').head()

data.plot(x='Event',y='Athlete',kind = 'bar', figsize = (10,5),color = 'yellow')
```
